```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
dataset = pd.read_csv('/content/w2_position_salaries-220925-153437.csv')
```

```python
X = dataset[["Level"]]   # must be 2D
y = dataset["Salary"]
lin_reg = LinearRegression()
lin_reg.fit(X, y)
y_pred_lin = lin_reg.predict(X)
r2_lin = r2_score(y, y_pred_lin)

print("Linear Regression R2 Score:", r2_lin)
```

```python
poly_2 = PolynomialFeatures(degree=2)
X_poly_2 = poly_2.fit_transform(X)

lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly_2, y)

y_pred_2 = lin_reg_2.predict(X_poly_2)
r2_poly_2 = r2_score(y, y_pred_2)

print("Polynomial Regression (Degree 2) R2 Score:", r2_poly_2)
```

```python
poly_3 = PolynomialFeatures(degree=3)
X_poly_3 = poly_3.fit_transform(X)

lin_reg_3 = LinearRegression()
lin_reg_3.fit(X_poly_3, y)

y_pred_3 = lin_reg_3.predict(X_poly_3)
r2_poly_3 = r2_score(y, y_pred_3)

print("Polynomial Regression (Degree 3) R2 Score:", r2_poly_3)
```

```python
poly_4 = PolynomialFeatures(degree=4)
X_poly_4 = poly_4.fit_transform(X)

lin_reg_4 = LinearRegression()
lin_reg_4.fit(X_poly_4, y)

y_pred_4 = lin_reg_4.predict(X_poly_4)
r2_poly_4 = r2_score(y, y_pred_4)

print("Polynomial Regression (Degree 4) R2 Score:", r2_poly_4)
```

```python
print("\nSummary:")
print("Linear Regression:", r2_lin)
print("Poly Degree 2:", r2_poly_2)
print("Poly Degree 3:", r2_poly_3)
print("Poly Degree 4:", r2_poly_4)
```
